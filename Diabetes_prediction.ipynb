Doing the Imputation

```python
df.columns
```

```python
df['Pregnancies']=df['Pregnancies'].replace(0,df['Pregnancies'].median())
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].mean())
df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].mean())
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].median())
df['BMI']=df['BMI'].replace(0,df['BMI'].mean())
df['DiabetesPedigreeFunction']=df['DiabetesPedigreeFunction'].replace(0,df['DiabetesPedigreeFunction'].median())
df['Age']=df['Age'].replace(0,df['Age'].median())
```

```python
df.head(10)
```

**Checking and fixing for outliers**

```python
## spliiting the data set x-> input feature and y -> target value
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
ax=sns.boxplot(data=x,ax=ax)
plt.savefig('boxplot.png')
```

```python
df.columns
```

```python
cols=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
for col in cols:
  Q1=x[col].quantile(0.25)
  Q3=x[col].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  mask=(x[col]>=lower_bound)&(x[col]<=upper_bound)
```

```python
x_outlier_detection =x[mask]
y_outlier_detection =y[mask]
```

```python
x_outlier_detection.shape
```

```python
y_outlier_detection.shape
```

**Using quantile method - removed more outliers from each columns**

```python
x_scaled.reset_index(drop=True, inplace=True)
y_outlier_detection.reset_index(drop=True, inplace=True)

q = x_scaled['Insulin'].quantile(0.95)
mask = x_scaled['Insulin'] < q
datanew = x_scaled[mask]
y_outlier_detection = y_outlier_detection[mask]
```

**Model Training**

splitting of data in testing training set

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(datanew,y_outlier_detection,test_size=0.33,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

data imbalancing

1.   over sampling -> minority class and increase that number to majority class

2.   under sampling -> majority class and increasing that number to majority class

2.   SMOTE : create a synthetic data and increase it to majority class





```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_train_smote,y_train_smote=smote.fit_resample(x_train,y_train)
```

```python
y_train_smote.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression()
classification.fit(x_train_smote,y_train_smote)
```

**Model Evaluation**

```python
y_pred=classification.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
target_names=['non diabetic','diabetic']
print(classification_report(y_test,y_pred,target_names=target_names))
```

**Saving the Model**
